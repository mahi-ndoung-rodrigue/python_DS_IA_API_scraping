````python
# Extract tables from webpage using Pandas. Retain table number 3 as the required dataframe.
# Extraire des tableaux de la page Web à l’aide de Pandas. Conservez la table numéro 3 comme trame de données requise.

tables = pd.read_html(URL)
df = tables[3]

# Replace the column headers with column numbers
df.columns = range(df.shape[1])



# Retain columns with index 0 and 2 (name of country and value of GDP quoted by IMF)
# Conserver les colonnes avec les indices 0 et 2 (nom du pays et valeur du PIB cité par le FMI)
df = df[[0,2]]



# Retain the Rows with index 1 to 10, indicating the top 10 economies of the world.
# Conserver les lignes avec l’indice 1 à 10, indiquant les 10 premières économies du monde.
df = df.iloc[1:11,:]



# Assign column names as "Country" and "GDP (Million USD)"
# Attribuez des noms de colonne comme « Pays » et « PIB (Million USD) »

df.columns = ['Country','GDP (Million USD)']
````
